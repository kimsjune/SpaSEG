# load the estimate dataset

```r
OvarianCancerExpr <- system.file("extdata", "sample_input.txt", package="estimate")
read.table(OvarianCancerExpr)
```

# Convert the expression matrix to a gct format file

```r
ESTIMATE_gct_path = "../../result/cci/IDC/Estimate_analysis/ESTIMATE_gct/"

ESTIMATE_score_path = "../../result/cci/IDC/Estimate_analysis/ESTIMATE_score/"

txt_path = '../../result/cci/IDC/trans_border_expr/'
```

### Our data: border_adata_expr_-4--8

```r
border_4_8 = read.table(paste0(txt_path,"trans_border_adata_expr_-4--8.txt"))
border_4_8
```

```r
in.gct.file_border_4_8 =  paste0(ESTIMATE_gct_path,'estimate_input_border_-4--8.gct')  # the gct file path to save

# Write gene expression data in GCT formatt
outputGCT(border_4_8, in.gct.file_border_4_8)
```

```r
# filterCommonGenes will get the common genes of expression matrix and estimate dataset

filterCommonGenes(input.f=paste0(txt_path,"trans_border_adata_expr_-4--8.txt"),       # Input file: expression matrix path
                  output.f=in.gct.file_border_4_8,        # the gct file path to save
                  id="GeneSymbol")                    #The row name of our data set is GeneSymbol
```

# Calculation of immune and matrix scores

```r
out.score.file_border_4_8 =  paste0(ESTIMATE_gct_path,"estimateScore_border_-4--8.gct")  # estimateScore save path

estimateScore(input.ds = in.gct.file_border_4_8, 
              output.ds=out.score.file_border_4_8, 
              platform="affymetrix") 
```

```r
scores=read.table(out.score.file_border_4_8,
                  skip = 2,
                  header = T,
                  row.names = 1 
                 )
View(scores)
```

```r
scores = as.data.frame(t(scores[,2:ncol(scores)]))
scores$Samples = rownames(scores)
scores = scores[,c(ncol(scores),2:ncol(scores)-1)]
```

```r
scores
```

```r
write.table(scores, file=paste(ESTIMATE_score_path,'IDC_Estimate_score_border_-4--8.csv',sep=""),
            quote=F,
            sep=',',
            row.names=F   
           )
```
