### Plot the ARI score for MB cellbin, seqFISH, MERFISH data

```r
library(ggplot2)
options(repr.plot.width = 4.5/2.54, repr.plot.height = 4.5/2.54, repr.plot.res = 400)
```

```r
ARI = c(0.14, 0.12, 0.11, 0.46, 0.38, 0.45, 0.45, 0.46, 0.26, 0.38, 0.33)
platform = c("MB cellbin","MB cellbin","MB cellbin", "seqFISH","seqFISH","seqFISH","seqFISH", "MERFISH","MERFISH","MERFISH","MERFISH")
method = c('SpaSEG','SpaGCN','Leiden','SpaSEG','SpaGCN','Leiden','BayesSpace','SpaSEG','SpaGCN','Leiden','BayesSpace')

df <- data.frame(method=method, platform=platform, ARI=ARI)

df$method <- as.factor(df$method)
df$method <- factor(df$method,levels = c('SpaGCN','BayesSpace','Leiden','SpaSEG'))
#df$method <- factor(df$method,levels = c('SpaSEG','SpaGCN','Leiden','BayesSpace'))


# df$platform <- as.factor(df$platform)
# df$platform <- factor(df$platform,levels = c('MB cellbin','seqFISH','MERFISH'))
```

```r
#options(repr.plot.width = 6/2.54, repr.plot.height = 4.5/2.54, repr.plot.res = 400)
options(repr.plot.width = 5.2/2.54, repr.plot.height = 3.3/2.54, repr.plot.res = 400)

p1 <- ggplot(df, aes(x=platform, y=ARI, fill = method)) +
  geom_bar(stat="identity", width=0.8,position = position_dodge(preserve = "single"), alpha = 0.75) +
  scale_fill_manual(values=c('#67C2A3','#8EA0C9','#FC8A61','#E889BD')) +
#   geom_text(aes(label = ARI), fontface = "plain", vjust=-0.2,angle=55,#vjust = 1.5,# angle = 45,
#              position = position_dodge(.9), size = 1.4, check_overlap = TRUE) + #
  scale_y_continuous(limits = c(0, 0.5), seq(0,0.5,by=0.1), name = "ARI") +
  theme_bw() + 
  #ylab("ARI") +
  theme(#legend.position =c(0.3, 0.65),
        legend.position =c(0.25,0.85),
        #legend.title = element_text(size=5),
        legend.text = element_text(size=5),
        legend.title = element_blank(),
        legend.key.size = unit(0.2, 'cm'),
        legend.margin=margin(0,0,0,0),
        legend.box.margin=margin(0,0,-10,-10),
        legend.spacing.x = unit(0.05, 'cm'),
        legend.spacing.y = unit(0, 'cm'),
        #panel.background = element_blank(),
        #panel.border = element_blank(),
        panel.grid = element_line(size = 0.15, color="#E2E2E2"), ##E2E2E2
#         panel.grid.major = element_line(size = 0.2, linetype = 'solid'), 
#         panel.grid.minor = element_line(size = 0.15, linetype = 'solid'),
        axis.text=element_text(colour="black",size=5, family="sans"),#设置坐标轴刻度数字
        #axis.text.x=element_text(angle = 35, hjust = 1),
        #axis.text.y = element_blank(),
        axis.line = element_blank(),
        #axis.title.y = element_blank(), 
        axis.title.x = element_blank(),
        axis.title.y = element_text(colour="black", size=5),
        axis.ticks.x=element_line(color="black",size=0.25,lineend = 1),
        axis.ticks.y=element_line(color="black",size=0.25,lineend = 1),
        axis.text.x=element_blank()
        #axis.ticks.y=element_blank()
#        axis.ticks.y=element_line(color="black",size=0.25,lineend = 1)
  )
p1
```
